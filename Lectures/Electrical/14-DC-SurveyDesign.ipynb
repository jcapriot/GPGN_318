## The measurement!
We measure $\Delta \phi$ and infer $\sigma$ ( or $\rho = 1/\sigma$)

If it was a halfspace then we could directly measure conductivity with our array:

$$
\Delta \phi = \frac{I}{2 \pi \sigma}\left(\frac{1}{R_{AM}} - \frac{1}{R_{BM}} - \frac{1}{R_{AN}} + \frac{1}{R_{BN}} \right)
$$

Or a bit more succinctly as:
$$ \Delta \phi = \frac{I}{\sigma} G $$
$$\text{with } G = \frac{1}{2\pi}\left(\frac{1}{R_{AM}} - \frac{1}{R_{BM}} - \frac{1}{R_{AN}} + \frac{1}{R_{BN}} \right)$$

Solving for an apparent $\sigma$
$$\sigma_{app} = \frac{I}{\Delta \phi} G$$

## Conductivity

In simplest form, all rocks are made up of the following general features
* Rock (solid minerals)
* Pores (non-rock)
* Fluids (water, oil, gas, etc)

Earths geology provides countless combinations of the above factors
* All rock types contain a different combination of minerals in different volumes
* Some rocks are non-porous, which does not allow for pore-fluids to be present
* Depending on scale, we can have all pore space, i.e. no rock, and we might call that a void or a cave
* Voids can be filled with air or water, making them either resistive or conductive
* Pore space can have a single fluid, such as water, or multiple fluids such and oil and gas
* Connectivity of pore-space effects pathway of currents (tortuosity).
* Etc.

This results in the broad range of resistivities we observe in practice

### Archies Law
Named after Gus Archie (1907  1978) is an empirical quantitative relationship between porosity, electrical conductivity, and the water saturation of a rock
$$ \rho = \rho_w \alpha \varphi^{-m}S_w^{-n}$$

* $\rho$ - resistivity of the water saturated rock (bulk resistivity)
* $\rho_w$ 	-	resistivity of the pore fluid (water)
* $\alpha$		-	tortuosity factor that depends on the lithology (typical range: 0.5  1.5)
* $\varphi$		-	porosity of the rock (between 0  1)
* $S$		-	water saturation for the rock (between 0  1)
* $m$		-	cementation factor (typical range: 1.8  2)
* $n$		-	saturation exponent (typically around 2)

**Note**: Archies law assumes rock matrix is non-conductive (good for sandstones). When clay is present, Archies Law breaks down, and there are other equations such as Waxman-Smits that try to correct for this

## But how does it change the data???

Q? What happens to the currents at interfaces between materials?

* We know :$\nabla \cdot \vec{J} = 0$ when no source or sink
* All currents entering a volume must leave that volume.
* At boundaries: All current entering one object must leave from another!
  $$ \vec{J}\cdot\hat{n} = \sigma_1\vec{E}_1\cdot\hat{n} = \sigma_2 \vec{E}_2 \cdot \hat{n}$$
  $$ \vec{E}_{n, 1} \neq \vec{E}_{n,2}$$

or:
$$\left( \frac{1}{\sigma_2} - \frac{1}{\sigma_1}\right) \vec{J}_n = (\rho_2 - \rho_1) \vec{J}_n = \frac{\tau}{\varepsilon_0}$$

* $\tau$ - surface charge density!

Secondary charges change the electrical potential at measurement locations!
$$\Delta V = V_M - V_N$$
