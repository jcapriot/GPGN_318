## Background

### Geophysics Discovery Lab:
Many targets have been buried beneath Kafadar Commons as part of the Mines Geophysics Discovery Lab (GDL). These targets contain a broad collection of features representing different categories of geophysical problems, different combinations of physical properties, and different geometries, depths, and depth extents. This includes several magnetic targets with varying shapes and levels of magnetization, including: full UXO grid, manhole cover, long dipping gas-line, M with different properties in each segment, overlapping pipes, coffins, metal-cased borehole, and a few rumored hidden targets. In this lab, you will select a feature buried at the Mines GDL, design and set-up an appropriate survey grid, and gather magnetic field data over it. Your instructor will work with you to select the targets and design your survey grids.

## Tasks
### Data Collection
Given the above prior information on the Mines GDL
1. Set up a GPS base station for your differential GPS, you should all share this base station.
2. Set up a Magnetics Base station using the Proton-precession magnetometer, apart from your surveys (likely on Stratton commons).
3. Split into teams and choose a target, and a magnetometer. Some good targets are:
   * UXO grid
   * Dipping gas line
   * Manhole Cover
   * Overlying Pipes
   * Borehole (Easy one to visually see as it's at the surface)
   * The "M"
4. Establish a survey grid appropriate for collecting magnetic data over a well-defined target. Items to focus on include grid size, line direction, line spacing, and station (marker) spacing.
